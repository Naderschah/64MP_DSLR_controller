```python
## Random Assignment
import matplotlib.pyplot as plt
import numpy as np

mass = np.array([91.1893, 91.1863, 91.1894, 91.1853, 91.1875])
x_std = np.array([0.0031, 0.0028, 0.0030, 0.0029, 0.0021])
width = np.array([2.4959, 2.4876, 2.5025, 2.4947, 2.4952])
y_std = np.array([0.0043, 0.0041, 0.0041, 0.0041, 0.0023])

colors = ['blue', 'red', 'green', 'orange', 'grey']
legend = ['ALEPH', 'DELPHI', 'L3', 'OPAL', 'LEP']

fig, ax = plt.subplots()
plt.grid()
ax.set_title("LEP mass and decay width data")

for i in range(len(mass)):
    ax.plot(mass[i], width[i], 'o', color=colors[i],label=legend[i])
    ellipse1 = plt.matplotlib.patches.Ellipse((mass[i], width[i]), width=x_std[i]*2, height=y_std[i]*2, color=colors[i], alpha=0.2)
    ellipse3 = plt.matplotlib.patches.Ellipse((mass[i], width[i]), width=x_std[i]*6, height=y_std[i]*6, color=colors[i], fill=False, linestyle='dashed')
    ax.add_patch(ellipse1)
    ax.add_patch(ellipse3)

ax.set_xlabel("$m_Z\ [GeV]$", fontsize=14)
ax.set_ylabel("$\Gamma_Z\ [GeV]$", fontsize=14)

# Adjust plot limits
ax.set_xlim(np.min(mass - x_std*3.5), np.max(mass + x_std*3.5))
ax.set_ylim(np.min(width - y_std*3.5), np.max(width + y_std*3.5))
plt.legend()
plt.show()
```
